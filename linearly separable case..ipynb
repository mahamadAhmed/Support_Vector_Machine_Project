### Importing needed libraries

```python
import numpy as np
import cvxopt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix
```

### Class SVM that implements svm from scratch using lagrange dual:

```python
class SVM:
    def fit(self, X, y):
        n_samples, n_features = X.shape

        #Prepare my Lagrange dual equation components to prepare them for solving
        #using cvxopt to handle matrix operations

        #P = X^T X --> the norm of w2 
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                K[i,j] = np.dot(X[i], X[j])
        P = cvxopt.matrix(np.outer(y, y) * K)
        #q = -1 
        q = cvxopt.matrix(np.ones(n_samples) * -1)

        # A = y^T 
        A = cvxopt.matrix(y, (1, n_samples))

        # b = 0 
        b = cvxopt.matrix(0.0)

        #G = -1 (NxN)
        G = cvxopt.matrix(np.diag(np.ones(n_samples) * -1))

        #H = 0 (1xN)
        h = cvxopt.matrix(np.zeros(n_samples))

        #Solving our lagrange quad equation using solver function
        solution = cvxopt.solvers.qp(P, q, G, h, A, b)

        #Lagrange multipliers (Alphas)
        a = np.ravel(solution['x'])

        # Lagrange have non zero lagrange multipliers
        sv = a > 1e-5
        ind = np.arange(len(a))[sv] #range of alphas where alpha>0
        self.a = a[sv]
        self.sv = X[sv]
        self.sv_y = y[sv]

        # Bias (b)
        self.b = 0
        for n in range(len(self.a)):
            self.b += self.sv_y[n]
            self.b -= np.sum(self.a * self.sv_y * K[ind[n], sv])
        self.b /= len(self.a)

        # Weights using sum(alpha[i]+Xn[i]+yn[i])
        self.w = np.zeros(n_features)
        for n in range(len(self.a)):
            self.w += self.a[n] * self.sv_y[n] * self.sv[n]
        
    def project(self, X):
        return np.dot(X, self.w) + self.b #-->classifer formual
        
    def predict(self, X):
        #Sign function to return -1 or 1
        return np.sign(self.project(X))
```

#### Creating our sample data

```python
#Generating Data points
X, y = make_blobs(n_samples=250, centers=2,
                  random_state=0, cluster_std=0.60)

#Assign 1 to labels with value 0
y[y == 0] = -1
tmp = np.ones(len(X))
y = tmp * y
```

#### Plotting and splitting the data

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Creating and training an instance of our support vector machine class.

```python
svm = SVM()
svm.fit(X_train, y_train)
```

### plottiing the decision boundary and support vectors.

```python
def f(x, w, b, c=0):
    return (-w[0] * x - b + c) / w[1]
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='winter')
# w.x + b = 0
a0 = -4; a1 = f(a0, svm.w, svm.b)
b0 = 4; b1 = f(b0, svm.w, svm.b)
plt.plot([a0,b0], [a1,b1], 'k')
# w.x + b = 1
a0 = -4; a1 = f(a0, svm.w, svm.b, 1)
b0 = 4; b1 = f(b0, svm.w, svm.b, 1)
plt.plot([a0,b0], [a1,b1], 'k--')
# w.x + b = -1
a0 = -4; a1 = f(a0, svm.w, svm.b, -1)
b0 = 4; b1 = f(b0, svm.w, svm.b, -1)
plt.plot([a0,b0], [a1,b1], 'k--')
```

### Using the model to predict target and test accuracy using confusion matrix

```python
#generalization error
y_pred = svm.predict(X_test)
confusion_matrix(y_test, y_pred)
```
